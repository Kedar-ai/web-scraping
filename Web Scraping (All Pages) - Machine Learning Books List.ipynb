```python
# Import the necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin
import warnings
warnings.filterwarnings('ignore')
```

```python
# Create an empty dataframe to store the data
columns = ['Book Name', 'Pages', 'Year', 'Book Size', 'Downloads', 'Book Link']
df = pd.DataFrame(columns=columns)
```

```python
# Start on the first page
page = 1

while True:
    # Make a request to the website
    url = f'https://www.pdfdrive.com/search?q=machine+learning&pagecount=&pubyear=&searchin=&page={page}'
    response = requests.get(url)
    link = 'https://www.pdfdrive.com'

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all the book elements on the page
    books = soup.find_all('div', class_='file-right')

    # Iterate through the books and extract the book name, pages, and downloads
    for book in books:
        book_name = book.h2.text
        try:
            book_pages = book.find('span',{'class':'fi-pagecount'}).text
        except:
            book_pages = 'None'
        try:
            book_year = book.find('span',{'class':'fi-year'}).text
        except:
            book_year = 'None'
        try:
            book_size = book.find('span',{'class':'fi-size hidemobile'}).text
        except:
            book_size = 'None'
        try:
            book_downloads = book.find('span',{'class':'fi-hit'}).text
        except:
            book_downloads = 'None'
        try:
            book_link =urljoin(link, book.a.get('href'))
        except:
            book_link = 'None'

        # Append the data to the dataframe
        df = df.append({'Book Name': book_name, 'Pages': book_pages, 'Year': book_year, 'Book Size': book_size, 'Downloads': book_downloads, 'Book Link': book_link}, ignore_index=True)

    # Check if there is a next page
    next_page = soup.find('a', class_='navigation next')
    if next_page is None:
        break
    else:
        page += 1
```

```python
# you can also save the dataframe to a csv file
df.to_csv('Machine Learning Books Data.csv')
```
